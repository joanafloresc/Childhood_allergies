# 05 Data Hypothesis

Importing Libraries.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

Reading file.

```python
data = pd.read_csv('../01_data/clean_data.csv').iloc[:,1:]
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
data.groupby('allerg_c_s').agg({'subject_id':'count'})#.iloc[1:,:]
```

### Number of allergies is significantly different from 1 in subjects with at least 1 allergy.

```python
hyp_pool_1 = data[ data['allerg_c_s'] >= 1 ]['allerg_c_s']
hyp_pool_1
```

```python
stats.ttest_1samp(hyp_pool_1, popmean = 1, alternative = "two-sided")
```

It is significantly different.

### Follow-up is in average superior to 5 years?

```python
data.head()
```

```python
data['follow_up']=data['age_end_years']-data['age_start_years']
data.describe()
```

```python
stats.ttest_1samp(data['follow_up'], popmean = 5, alternative = "less")
```

In average, follow-up is superior to 5 years.

### Is the number of allergies at the start of the follow-up similar to the number at the end?

```python
stats.ttest_ind(data['allerg_c_s'], data['allerg_c_e'])
```

Null hypothesis is rejected, samples have different means,as such, mean number of allergies at the start is not similar to the mean number at the end. (Confidence Interval = 99% (p-value < 0.01 ))
